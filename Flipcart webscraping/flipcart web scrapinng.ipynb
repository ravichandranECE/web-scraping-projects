### flipcart web scraping

```python

import requests
from bs4 import BeautifulSoup
import pandas as pd
names=[]
prices=[]
descrip=[]
review=[]


for i in range(1,9):
    url="https://www.flipkart.com/search?q=mobiles%20under%2050000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"+str(i)
    r=requests.get(url)
    soup=BeautifulSoup(r.text,"lxml")
    box=soup.find("div",class_="_1YokD2 _3Mn1Gg")# these div using the select the particular content of page


    name=box.find_all("div",class_="_4rR01T")# box is the exctract the data from particularly
    for i in name:
        n=i.text
        names.append(n)
    print(len(names))
    
    price=box.find_all("div",class_="_30jeq3 _1_WHN1")
    for i in price:
        p=i.text
        prices.append(p)
    print(len(prices))
    
    decs=box.find_all("ul",class_="_1xgFaf")
    for i in decs:
        d=i.text
        descrip.append(d)
    print(len(descrip))
    
    rew=box.find_all("div",class_="_3LWZlK")
    for i in rew:
        reviwes=i.text
        review.append(reviwes)
    print(len(review))
```

```python

df=pd.DataFrame({"proudct name":names,"prices":prices,"description":descrip,"reviews":review})
df
```
